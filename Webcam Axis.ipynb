```python
import cv2
cap=cv2.VideoCapture(0)
while True:
    ret,frame=cap.read()
    cv2.imshow('our live sketch',frame)
    if cv2.waitKey(1)==13:
        break
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import matplotlib.pyplot as plt
cap=cv2.VideoCapture(0)
if cap.isOpened():
    ret,frame=cap.read()
    print(ret)
    print(frame)
img1=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
plt.imshow(img1)
plt.title("capture Image")
plt.xticks([])
plt.yticks([])
plt.show()
cap.release()

```
